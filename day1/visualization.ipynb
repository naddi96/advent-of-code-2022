```python
import matplotlib.pyplot as plt
with open("input_problem1.txt","r") as f:
    li=[]
    for i in f.readlines():
        i=i.replace("\n","")    
        if len(i)>1:
            li.append(int(i))



plt.plot(sorted(li))
plt.rcParams["figure.figsize"] = (20,20)
plt.show()
```

```python
plt.hist(li, density=1, bins=100) 
plt.show()
```
